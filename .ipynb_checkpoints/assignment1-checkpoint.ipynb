## Examine data entry errors

## Task 2.1

### Prepare DataFrame for 2014 data

```python
# Extract rainfall data in 2014
raw_rainfall_2014 = rainfall_df.loc[rainfall_df['Year'] == 2014]

# Reshape the data frame as required
rainfall_2014 = raw_rainfall_2014.pivot(index='Day', columns='Month', values='Rainfall amount (millimetres)')
rainfall_2014
```

### Explore the highest daily rainfall each month


```python
# Find the highest rainfall amount in each month 
highest_rainfall_values = rainfall_2014.max(axis=0, skipna=True)
# Find the day with the highest rainfall 
highest_rainfall_days = rainfall_2014.idxmax(axis=0)
```

```python
# Create a DataFrame with the highest rainfall days
highest_rainfall_2014 = pd.DataFrame({'Highest rainfall day': highest_rainfall_days})

# Calculate average monthly rainfall
total_monthly_rain_2014 = rainfall_2014.sum(axis=0)

# Join the DataFrame with the corresponding Series
highest_rainfall_2014 = highest_rainfall_2014.join(highest_rainfall_values.rename('Rainfall amount'))

# Join the average monthly rainfall 
highest_rainfall_2014 = highest_rainfall_2014.join(total_monthly_rain_2014.rename('Total monthly rain'))
highest_rainfall_2014
```

```python
# Create a grouped bar chart to compare the highest daily 
# rainfall for each month, and total monthly rainfall 

x = np.arange(len(highest_rainfall_2014.index))
width = 0.3 

plt.figure(figsize=(10, 7))
plt.bar(x - width/2, highest_rainfall_2014['Rainfall amount'], width, label='Highest Daily Rainfall', color='lemonchiffon')
plt.bar(x + width/2, highest_rainfall_2014['Total monthly rain'], width, label='Total Monthly Rainfall', color='paleturquoise')

plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.title('Highest Daily Rainfall By Month')
plt.xticks(x, highest_rainfall_2014.index)
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
highest_rainfall_2014.describe()
```

### Observations


1. The highest daily rainfall typically occurs in the beginning and the end of each month, with the highest rain data recorded on November 16th with 36.6mm.
 
2. The highest daily rainfall indeed does not represent the trend of rainfall throughout the year, but we can identify the rainfall distribution throughout the month. In particular, the total rainfall in April is exceptionally high in the year, but the highest daily rainfall recorded is only one-fourth of the total. This indicates that this month witnesses multiple number of rainy days. 

3. The mean rainfall for the highest daily value is 13mm, whereas the mean total monthly rainfall is 36mm. This indicates that the highest daily rainfall contributes about one-third of the total monthly rainfall.



## Task 2.2

```python
# Function to extract total monthly rainfall 
def monthly_rainfall(df, year):
    return df.loc[df['Year'] == year].groupby('Month')['Rainfall amount (millimetres)']
```

```python
# Extract rainfall data from 2015 to 2017
rainfall_2015 = rainfall_df.loc[rainfall_df['Year'] == 2015]
rainfall_2016 = rainfall_df.loc[rainfall_df['Year'] == 2016]
rainfall_2017 = rainfall_df.loc[rainfall_df['Year'] == 2017] 

# Calculate monthly rainfall for each year
monthly_rain_2015 = monthly_rainfall(rainfall_df, 2015).sum()
monthly_rain_2016 = monthly_rainfall(rainfall_df, 2016).sum()
monthly_rain_2017 = monthly_rainfall(rainfall_df, 2017).sum()


# Create a dataframe containing the monthly rainfall for 3 years
monthly_rain = pd.DataFrame({
    '2015': monthly_rain_2015,
    '2016': monthly_rain_2016,
    '2017': monthly_rain_2017
})
```

### Yearly rainfall trend

```python
"""
Plot the overall trend with average rainfall in the period
"""
# Extract yearly rainfall in 2015 - 2017
total_rain_2015_2017 = yearly_rainfall.loc[2015:2017]

# Plot the trend
ax = total_rain_2015_2017.reset_index().plot.bar(x='Year', y='Rainfall amount (millimetres)', figsize=(10,7), rot=0, color='cadetblue')
plt.title('Precipitation in 2015 - 2017')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')

plt.show()
```

### Observations
Overall, the period witness an increasing trend in the total annual rainfall amount. In 2015, the annual rainfall is about 400mm, which is much lower compared to the counterparts. This increasing trend indicates a shift in the climate conditions. 

```python
"""
Line graph: Show monthly rainfall trend in each year 
"""

rain_lines = monthly_rain.plot.line(figsize=(10,7))
plt.ylabel('Total rainfall (mm)')
plt.title('Monthly precipitation distribution by year')
plt.show()
```

#### Observations
- Based on the total annual rainfall comparison in the bar chart above, it can be seen that 2016 and 2017 share similar total rainfall figures; nevertheless, they have noticeably distinct rainfall distribution throughout the year. 
- 2017 witnessed highest rainfall levels in April and December with the amount exceeding 120mm. These anomalous figures could indicate extreme events in the recorded months. Also, these data points made a significant contribution to the overall annual rainfall. 
- In contrast, the year 2016 had a relatively stable rainfall distribution throughout the year, with the exception of a dip in February, recognized as the driest month. 
- The rainfall trend in 2015 is on a downward pattern, keeping the monthly records below 60mm, and reaching a trough of less than 20mm in October. 

### Average monthly rainfall in three-year period

```python
ax = monthly_rain.mean(axis=1).plot.bar(figsize=(10,7), rot=0, color='cadetblue')
plt.title('Average monthly rainfall in the period 2015-2017')
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm)')

plt.axhline(monthly_rain.mean(axis=1).mean(), label='Average monthly rainfall')
plt.legend()
plt.show()
```

### Observation
- The high levels of mean rainfall in April and December are largely contributed by the records in 2017. 
- In comparison, the monthly rainfally has a tendency to increase from July to October. This could indicate the rainy season in the city.
- The total monthly rainfally is generally recorded to be at low amount for March and June, indicating these months as the driest in the period. 

## Task 2.3

### Highest 3 years vs Lowest 3 years

```python
# Sort the year from highest to lowest rainfall
sorted_yearly_rainfall = yearly_rainfall.reset_index().sort_values(by=['Rainfall amount (millimetres)'], ascending=False, inplace=False)

# Extract top 3 years
top_highest = sorted_yearly_rainfall.head(3)
top_lowest = sorted_yearly_rainfall.tail(3)

fig, axs = plt.subplots(1, 2, figsize=(12,6))

# Highest rainfall years
axs[0].bar(top_highest['Year'].astype(str), top_highest['Rainfall amount (millimetres)'],color='lightseagreen')
axs[0].set_title('Top 3 highest rainfall years')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Rainfall (mm)')

# Lowest rainfall years
axs[1].bar(top_lowest['Year'].astype(str), top_lowest['Rainfall amount (millimetres)'], color='lightcoral')
axs[1].set_title('Top 3 lowest rainfall years')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Rainfall (mm)')

plt.tight_layout()
plt.show()
```

### Observations

- Overall, the total rainfall in top 3 lowest year is half the amount of top 3 highest year.
- The period of 2019-2021 witnessed a significant rise in the total annual rainfall, indicating a shift in the climate conditions.
- The period of 2013-2014 could be signified as the driest period in the inspected courses of year, as the total annual rainfall is below 450mm.


### Rainfall distribution over year

```python
# Extract total monthly rainfall for top 3 lowest years
monthly_rain_2014 = monthly_rainfall(rainfall_df, 2014).mean()
monthly_rain_2013 = monthly_rainfall(rainfall_df, 2013).mean()
monthly_rain_2019 = monthly_rainfall(rainfall_df, 2019).mean()

# Extract total monthly rainfall for top 3 highest years
monthly_rain_2020 = monthly_rainfall(rainfall_df, 2020).mean()
monthly_rain_2021 = monthly_rainfall(rainfall_df, 2021).mean()
monthly_rain_2022 = monthly_rainfall(rainfall_df, 2022).mean()

monthly_rain_lowest = pd.DataFrame({
    '2013': monthly_rain_2013,
    '2014': monthly_rain_2014,
    '2019': monthly_rain_2019,  
})
monthly_rain_highest = pd.DataFrame({
    '2020': monthly_rain_2020,
    '2021': monthly_rain_2021,
    '2022': monthly_rain_2022
})
```

```python
# Plot line graph

fig, axs = plt.subplots(2, 1, figsize=(10,6))

# Highest rainfall years
axs[0].plot(monthly_rain_lowest.index, monthly_rain_lowest['2013'],color='b', marker='o', label='2013')
axs[0].plot(monthly_rain_lowest.index, monthly_rain_lowest['2014'],color='g', marker='s', label='2014')
axs[0].plot(monthly_rain_lowest.index, monthly_rain_lowest['2019'],color='r', marker='^', label='2019')
axs[0].set_title('Top 3 lowest years - Monthly rainfall distribution')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Rainfall (mm)')
axs[0].legend()

# Lowest rainfall years
axs[1].plot(monthly_rain_highest.index, monthly_rain_highest['2020'],color='c', marker='o', label='2020')
axs[1].plot(monthly_rain_highest.index, monthly_rain_highest['2021'],color='m', marker='s', label='2021')
axs[1].plot(monthly_rain_highest.index, monthly_rain_highest['2022'],color='y', marker='^', label='2022')
axs[1].set_title('Top 3 highest years - Monthly rainfall distribution')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Rainfall (mm)')
axs[1].legend()

plt.tight_layout()
plt.show()
```

### Observations

Regarding the top 3 lowest years:
- We cannot conclude that 2013 is the driest year in the period, since its records just started from June. Nevertheless, the average monthly rainfall distribution of later half of this year is stably higher than its counterparts.
- 2019 experienced an intense variations of rainfall, since there were many months when the average monthly rainfall was below 0.5mm. The period from May to September saw a stable rise in the rainfall, however, the rise was minor.
- The annual rainfall of 2014 could be largely attributed to the peak in April, with the average rain of nearly 2.5mm. 
- By comparison, the period from January to March was the driest months, with the valley in January in both 2014 and 2019. The monthly rainfall had a tendency to increase in November, and then fell dramatically in December. 

For the top 3 highest years:
- By comparison, the monthly rainfall distribution of the 3 years shared relatively similar patterns. The rise averagely monthly rainfall could be expected in January and October, with a peak at nearly 5mm. The drop in rainfall could be seen in February, July and December, with the valley in February at nearly 0mm. The period from June to August had a low rainfall in 3 years. 
- 2020 witnessed the highest average monthly rainfall in April, which was unusual compared to its counterparts.
