## Practical exericise 1: Data retrieving

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
source = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
labels = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
abalone_df = pd.read_csv(source, sep=',', decimal='.', names=labels)
abalone_df.head()
```

```python
# Access the 4th line of Sex column
abalone_df.loc[3, 'Sex']
```

```python
# Access the 4th line of Sex column - Method 2
abalone_df.iloc[3,0]
```

### Using Mask

```python
# Check the diameter information of the abalone whose rings are 21
mask_ring21 = abalone_df['Rings'] == 21
abalone_df.loc[mask_ring21, 'Diameter']
```

### Grouping feature

```python
# Check the mean  value when grouping abalone by Rings
# Alternative: abalone_df.groupby(['Sex']).mean(numeric_only=True)
abalone_without_sex = abalone_df.drop(['Sex'], axis='columns', inplace=False)
abalone_without_sex.groupby(['Rings']).mean()
```

## Practical exercise 3: Basic operation

```python
# Sum of Shell weight
abalone_df['Shell weight'].sum()
```

```python
# Sort by Length in descending order
sorted_abalone = abalone_df.sort_values(by='Length', ascending=False)
sorted_abalone
```

```python
# Write to csv file
file_path = 'w4_processed_abalone.csv'
sorted_abalone.to_csv(file_path, index=False)
```

## Practical exercise 4: Visualization

### Pie chart

```python
abalone_df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
plt.show()
```

### Histogram

**Question: How can I plot a histogram if there is an outlier that far exceeds the normal values?

```python
abalone_df['Height'].plot(kind='hist', bins=20)
plt.xlabel('Height')
plt.show()
```

### Density

```python
abalone_df['Shell weight'].plot(kind='density')
plt.title('Abalone - Shell weight')
plt.xlabel('Shell weight')
plt.show()
```

### Boxplot

```python
abalone_df['Whole weight'].plot(kind='box')
plt.show()
```

```python
# Boxplot for Whole weight per Sex group
abalone_df.dropna().boxplot(column='Whole weight', by='Sex')
plt.show()
```

### Scatterplot

```python
abalone_df.plot(kind='scatter', x=8, y=2)
plt.show()
```

```python
# Scatterplot between Weight summation and Whole weight
abalone_df['Weight summation'] = abalone_df['Shucked weight'] + \
                                abalone_df['Viscera weight'] + abalone_df['Shell weight']
abalone_df.plot(kind='scatter', x='Weight summation', y='Whole weight')
plt.title('Abalone - Summation of Weights')
plt.xlabel('Weight summation')
plt.show()
```

```python
# Scatter Matrix
from pandas.plotting import scatter_matrix
scatter_matrix(abalone_df, alpha=0.2, figsize=(16,16), diagonal='hist')
plt.show()
```
